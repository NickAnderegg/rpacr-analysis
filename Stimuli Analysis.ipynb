```python
import pathlib
import csv, json
from pprint import pprint
```

# Load raw stimuli

```python
raw_sets = pathlib.Path('stimuli/revised_stimuli_2.csv')
with raw_sets.open('r') as f:
    csvreader = csv.reader(f, lineterminator='\n', delimiter='\t')
    
    stim_sets = dict()
    
    curr_dict = dict()
    curr_set = set()
    
    all_chars = set()
    
    for row in csvreader:
#         print(row)
        if not row:
            stim_sets[frozenset(curr_set)] = curr_dict
            curr_dict = dict()
            curr_set = set()
        elif row[0] == 'Critical:':
            curr_set.add(row[1])
            all_chars.add(row[1])
            curr_dict['critical'] = {
                'character': row[1],
                'phono': row[2],
                'log_count': float(row[4])
            }
            
            curr_dict['agg_score'] = float(row[3].split()[1])
        else:
            item = row[0][:-1].replace(' ', '_').lower()
            all_chars.add(row[1])

            curr_set.add(row[1])
            curr_dict[item] = {
                'character': row[1],
                'phono': row[2],
                'score': float(row[3].split()[1]),
                'log_count': float(row[4])
            }
            
raw_sets = stim_sets

pprint(raw_sets)
```

# Load trials

```python
trials_file = pathlib.Path('stimuli/trials.json')
with trials_file.open('r') as f:
    trials = json.load(f)['sentences']
```

# Isolate stimulus sets

```python
print(list(raw_sets.keys())[0])
trial_chars = set()
invalid_sets = 0
for trial in trials:
    if 'original_distractors' in trial:
        key = 'original_distractors'
    else:
        key = 'distractors'
    
    char_set = frozenset([trial[key]['both_sim'], trial[key]['orth_sim'], trial[key]['phon_sim'], trial['critical_target']])
    
    # trial_chars = trial_chars | char_set
    for char in char_set:
        if char not in all_chars:
            invalid_sets += 1
            break
            print(char, 'not in all_chars')
            
print(invalid_sets)
```
